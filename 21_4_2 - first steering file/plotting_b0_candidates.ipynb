# Plotting the distribution of the beam-constrained mass for the B0 candidates

```python
import root_pandas
import matplotlib.pyplot as plt
# show plots in notebook
%matplotlib inline
# set cool colours
plt.style.use('belle2')
```

## Signal + Background

Import the results obtained by `b0_candidates` for the filenumber `filenumber`

```python
# file to plot
filenumber = 1
filename = f"./Bd2JpsiKS_{filenumber}_without_cut.root"
# import to python
df = root_pandas.read_root(filename)
# Investigate ROOT file
df.describe()
```

Plot the distribution of the candidates taking advantage that this is MonteCarlo data and we know which candidates are actually `signal` (actual B0s) and which ones are `background`

```python
# Plot histogram in 100 bins
df.Mbc.hist(range=(4.3, 5.3), bins=100, by=df.isSignal)
plt.xlabel(r'M$_{\rm bc}$ [GeV/c$^{2}$]')
plt.ylabel('Number of candidates')
```

## Signal without Background

After adding a cut to remove the background, we can run everything again to see only the data we want

```python
filenumber = 1
filename = f"./Bd2JpsiKS_{filenumber}_with_cut.root"
# import to python
df2 = root_pandas.read_root(filename)
# Investigate ROOT file
df2
```

```python
# Plot histogram in 100 bins
df2.Mbc.hist(range=(4.3, 5.3), bins=100)
plt.xlabel(r'M$_{\rm bc}$ [GeV/c$^{2}$]')
plt.ylabel('Number of candidates')
```

# Plot ROE

```python
m_bins = 50
m_range = (0, 10)
fig, ax = plt.subplots(1,2, figsize=(15, 7))
# Left subplot of ROE mass:
ax[0].hist(df2['roeM__bo__bc'], label='No mask',
           bins = m_bins, range=m_range, alpha=0.6)
ax[0].hist(df2['roeM__bomy_mask__bc'], label='"my_mask" applied',
           bins = m_bins, range=m_range, alpha=0.6)
ax[0].set_xlim(m_range)
ax[0].set_xlabel('ROE mass [GeV/$c^2$]')
# Right subplot of number of charged ROE particles:
m_bins = 15
m_range = (0, 15)
ax[1].hist(df2['nROE_Charged__bo__bc'], label='No mask',
           bins = m_bins, range=m_range, alpha=0.6)
ax[1].hist(df2['nROE_Charged__bomy_mask__bc'], label='"my_mask" applied',
           bins = m_bins, range=m_range, alpha=0.6)
ax[1].set_xlim(m_range)
ax[1].set_xlabel('# of charged ROE particles')
ax[1].legend()
fig.tight_layout()
```

The invariant mass distribution for masked ROE is much narrower and its mean is a little bit below nominal $B^0$ mass, contrary to the unmasked ROE distribution. This is expected, because a generic $B^0$ decay may produce particles, that are not accounted in ROE mass computation, like $K^0_L$ or neutrinos.

The distribution of the number of charged particles for masked ROE has much more prominent peaks at 4 and 6 particles than its unmasked version, which corresponds to the fact that a correctly reconstructed $B^0$ will have an even number of charged daughter particles.

This means that even a simple ROE mask like my_mask does a really good job of cleaning up the particles, which are not associated to the partner $B^0$.
