# Plotting the distribution of the beam-constrained mass for the B0 candidates

```python
import root_pandas
import matplotlib.pyplot as plt
# show plots in notebook
%matplotlib inline
```

## Signal + Background

Import the results obtained by `b0_candidates` for the filenumber `filenumber`

```python
# file to plot
filenumber = 1
filename = f"./Bd2JpsiKS_{filenumber}_without_cut.root"
# import to python
df = root_pandas.read_root(filename)
# Investigate ROOT file
df.describe()
```

Plot the distribution of the candidates taking advantage that this is MonteCarlo data and we know which candidates are actually `signal` (actual B0s) and which ones are `background`

```python
# Plot histogram in 100 bins
df.Mbc.hist(range=(4.3, 5.3), bins=100, by=df.isSignal)
plt.xlabel(r'M$_{\rm bc}$ [GeV/c$^{2}$]')
plt.ylabel('Number of candidates')
```

## Signal without Background

After adding a cut to remove the background, we can run everything again to see only the data we want

```python
filenumber = 1
filename = f"./Bd2JpsiKS_{filenumber}_with_cut.root"
# import to python
df2 = root_pandas.read_root(filename)
# Investigate ROOT file
df2
```
